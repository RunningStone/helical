```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
def parse_string_to_array(s):
    # Remove brackets and decimal points
    s = s.strip('[]').replace('.', '')
    # Split by spaces and convert to integers
    numbers = [float(x) for x in s.split() if x]
    return np.array(numbers)
```

```python
train = train.map(lambda x: {'Value': parse_string_to_array(x['Value'])})
```

```python
test = test.map(lambda x: {'Value': parse_string_to_array(x['Value'])})
```
